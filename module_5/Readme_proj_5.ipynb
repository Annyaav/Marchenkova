### Проблемы, с которыми столкнулась в ходе работы.
1. Изначально было непонятно, как приступить к работе. В юните даны практически все подсказки, необходимые для выполнения проекта, однако собрать их в логическом порядке, организовать потоки данных - это была и есть основная проблема: видимо за счет недостаточного опыта.
2. Разделение на тестовую, обучающую и валидационную выборки и работа с ними. Этот момент достаточно слабо освещен (именно в части применения к данному проекту) поэтому делала интуитивно, и сейчас не уверена, что сделал правильно, хотя результат получен.

### Оценка результата
Конечный результат оцениваю как неудовлетворительный, поскольку Модель, которая выдает такие показатели точности и полноты, невозможно использовать не практике: слишком много ошибок.
По-видимому, для соревнования такой набор исходных данных подходит, однако для построения рабочей модели нужно больше данных: как расширение признакового пространства, так и увеличение временного охвата выборки.



### Вывод
Итак, более-менее можем считать рабочими три последних модели, несмотря на то, что показатель ROC AUC приблизительно одинаковый. 
Пока классы не сбалансированы в обучающей выборке, ни о какой достоверной классификации речи идти не может, несмотря на, казалось бы высокий, показатель Accuracy.
Но и в конечной модели показатель точности (т.е. всего 23% оценок достоверные) очень низкий.

Accuracy вообще не стоит использовать при ощутимой разбалансировке классов.
При оценке качества модели кроме ROC AUC (выбран как показатель качества модели в соревновании на Kaggle) Следует использовать такие стандартные показатели, как точность, полнота, F-мера. 
То есть только комплексная оценка даст достоверную информацию о возможности промышленного использования модели.

### Что не успела/не смогла сделать в рамках этого проекта.
1. Добиться желаемых показателей точности (80%). Очевидно, что для этого нужно было расширить признаковое пространство, но больше придумать не смогла на основании имеющихся данных. А может быть этот показатель для данной задачи вообще недостижим? Слишком много так называемого "человеческого фактора".
2. Также не нашла полной информации по расшифровке и значению гиперпараметров логистической регресси.
3. Не успела разобраться с автотюнингом модели, построенной  на LogisticRagression, думаю, эта возможность где-то есть в sklearn но не хватило времени поискать, в т.ч. и готовые решения. 
4. Не поробовала другие ML-алгоритмы, было бы интересно сравнить рзультаты работы разных алгоритмов для одной и той же задачи.
5. Категорически не нравятся мои графики в проекте. Процесс ML можно очень красиво и наглядно визуализировать, в т.ч. и графическое изображение недоученности/переученности модели, например.
Но для рисования красивых и разнообразных графиков катастрофически не хватает материалов, данных в курсе, и практики. Самостоятельно ищу и интернете, но это лишь обрывочные сведения, которые опять же, слложно упорядочить и применять самостоятельно.
6. Не удалось поучаствовать в командной работе по объективным причинам. Как житель Дальневосточного региона имею +7 часов к московскому времени. С бОльшим количество потенциальных участников команд я нахожусь в рабочей противофазе.
